Some fancy math about brownian motion

```python
# Adding module folder to path
import sys
import os
# Importing absolute paths
from Modules import modules_path

# Adding paths to the system paths
if str(modules_path) not in sys.path:
    sys.path.append(str(modules_path))

# Importing GBMSimulator class
from GBMSimulator import  GBMSimulator
import matplotlib.pyplot as plt
```

```python
# Parameters for GBM

'''
    Args:
        mu (int): drift coefficient
        n (int): number of steps
        T (int): time in years
        M (int): number of sims
        y0 (int): initial value
        sigma (int): volatility
    
'''
mu = 0.1
n = 100
T =  1
M = 1000
y0 = 0.01
sigma = 0.5
```

```python
# Initialize simulator
simulator = GBMSimulator()
```

```python
# Creating an array of simulated paths

simulated_paths = simulator.simulate(mu, n, T, M, y0, sigma)
```

```python
# template
plt.style.use('ggplot') 
# changing size of chart
plt.figure(figsize=(12, 5))

# Loop to do a line plot for each simulation
for num, sim in enumerate(simulated_paths.T):
    # length of each simulation
    t_values = list(range(len(sim)))
    # simulated values
    y_values = sim

    # line plot for each sim
    plt.plot(t_values, y_values, label=f"GBM path {num+1}")


# labelling chart
plt.xlabel("Time")
plt.ylabel("Y(t)")
plt.title("Simulated Geometric Brownian Motion Path")
if M <= 10:
    plt.legend()
else:
    plt.show()
```
